```python
import matplotlib.pyplot as plt
```

```python
#Creating a Bar Graph
x_axis = df['Position']
y_axis = df['Streams']

plt.bar(x_axis, y_axis)
plt.title('Spotify Dataset')
plt.xlabel('Position of Song')
plt.ylabel('Number of Streams')
plt.show()
```

The graph depicts the relationship between the position of the song on the leaderboard and the number of streams. We can see that there is a large difference between the number of streams of the song at position 1 with the number of streams for the song at the last position.
